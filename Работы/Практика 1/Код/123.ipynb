```python
#Импорт необходимых библиотек
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
#Создание случайных данных для обучения модели
raw_data = make_blobs(
    n_samples = 200, 
    n_features = 2, 
    centers = 4, 
    cluster_std = 1.8
)
```

```python
#Проверка корректности введенных данных
plt.scatter(raw_data[0][:,0], raw_data[0][:,1], c=raw_data[1]);
```

```python
#Обучение модели
model = KMeans(n_clusters=4)
model.fit(raw_data[0])
#Информация о принадлежности к определенному кластеру
model.labels_
```

```python
#Определение центров кластеров
model.cluster_centers_
```

```python
#Визуализация точности предсказания нашей модели
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('Наши предсказания')
ax1.scatter(raw_data[0][:,0], raw_data[0][:,1],c=model.labels_)
ax2.set_title('Реальные значения')
ax2.scatter(raw_data[0][:,0], raw_data[0][:,1],c=raw_data[1]);
```
